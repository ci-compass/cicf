# Visualizing Map-Reduce Results

This notebook visualizes the output of our Map-Reduce computation on LIGO data. 

### Objectives
1. Load the aggregated results from `map_reduce_report.json`.
2. Compare statistics across different LIGO data files.
3. Understand how the 'Reduce' step consolidated information from multiple sources.

```python
import json
import pandas as pd
import matplotlib.pyplot as plt
import os

# Ensure the report file exists
report_path = 'map_reduce_report.json'

if not os.path.exists(report_path):
    print(f"Error: {report_path} not found. Did you run 'python map_reduce_ligo.py' first?")
else:
    with open(report_path, 'r') as f:
        report_data = json.load(f)
    
    print(f"Successfully loaded report: {report_data['total_files_processed']} files processed.")
```

### 1. Data Processing
We will convert the `all_stats` list into a Pandas DataFrame for easier manipulation.

```python
df = pd.DataFrame(report_data['all_stats'])
df
```

### 2. Visualization: Max Strain per File
We'll use a bar chart to compare the maximum absolute strain detected in each file.

```python
plt.figure(figsize=(10, 6))
plt.bar(df['filename'], df['max_strain'], color='skyblue')
plt.xlabel('LIGO Data File')
plt.ylabel('Max Absolute Strain')
plt.title('Comparison of Max Strain Across Files')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

### 3. Visualization: Standard Deviation Comparison
The standard deviation gives us an idea of the 'noise' level in the data.

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['gps_start'], df['std_strain'], s=100, color='coral', marker='D')
plt.axhline(y=report_data['average_std_dev'], color='red', linestyle='--', label='Global Average')

for i, txt in enumerate(df['filename']):
    plt.annotate(txt, (df['gps_start'][i], df['std_strain'][i]), fontsize=8, xytext=(5,5), textcoords='offset points')

plt.xlabel('GPS Start Time')
plt.ylabel('Strain Standard Deviation')
plt.title('Noise Levels vs Observation Time')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

### Summary of Global Results
The Reduce step aggregated these findings into global constants:

```python
print(f"Global Max Strain: {report_data['global_max_strain']}")
print(f"Found in file: {report_data['global_max_strain_file']}")
print(f"Global Average Noise (Std Dev): {report_data['average_std_dev']}")
```
